```python
# factorize we use to convert string type data into numeric form.BY using factorize we can use index no.s to represnt data from that column
```

```python
def Encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate','cost','votes'])]:
        zomato[column]=zomato[column].factorize()[0]
    return zomato
```
